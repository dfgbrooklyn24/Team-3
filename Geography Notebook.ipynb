```python
# Filter the DataFrame to exclude rows with ZIP code 19140
filtered_data = phl_philadelphia_data[phl_philadelphia_data['ZIP'] != 19140]

# Display the filtered DataFrame
filtered_data
```

```python
grouped_df = filtered_data.groupby('ZIP').agg({
    'TRACT': lambda x: list(x),
    'RES_RATIO': lambda x: list(x),
    'BUS_RATIO': lambda x: list(x),
    'OTH_RATIO': lambda x: list(x),
    'TOT_RATIO': lambda x: list(x),
}).reset_index()
```

```python
import numpy as np

# Define a function to calculate the mean while handling non-numeric values
def calculate_mean(x):
    # Convert each element to float, ignoring non-numeric values
    numeric_values = [float(val) for val in x if val.replace('.', '').isdigit()]
    # Calculate the mean
    return np.mean(numeric_values)

# Apply the function to calculate the mean for each group
grouped_df['RES_RATIO'] = grouped_df['RES_RATIO'].apply(calculate_mean)
grouped_df['BUS_RATIO'] = grouped_df['BUS_RATIO'].apply(calculate_mean)
grouped_df['OTH_RATIO'] = grouped_df['OTH_RATIO'].apply(calculate_mean)
grouped_df['TOT_RATIO'] = grouped_df['TOT_RATIO'].apply(calculate_mean)
```

```python
grouped_df
```

```python
res_ratio_describe = grouped_df.groupby('ZIP')['RES_RATIO'].describe()

res_ratio_describe
```

```python
desired_zip_code_describe = desired_zip_code.describe()

desired_zip_code_summary = desired_zip_code_describe['RES_RATIO']

# Calculate the Euclidean distance between the desired ZIP code and all other ZIP codes
distances = {}
for zip_code, summary in res_ratio_describe.iterrows():
    if zip_code != desired_zip_code:
        distances[zip_code] = np.linalg.norm(desired_zip_code_summary - summary)

# Sort the ZIP codes based on their distance from the desired ZIP code
sorted_distances = sorted(distances.items(), key=lambda x: x[1])

# Select the top 10 ZIP codes with the smallest distances
top_10_related_zip_codes = sorted_distances[:10]

print("Top 10 ZIP codes closely related to", desired_zip_code, "based on 'RES_RATIO' column summary statistics:")
for zip_code, dist in top_10_related_zip_codes:
    print("ZIP Code:", zip_code, "| Distance:", dist)
```
